```python
#Data Preprocessing
```

```python
#Handling duplicate values
```

```python
df_olympics.duplicated().sum()
```

```python
df_olympics=df_olympics.drop_duplicates()
df_olympics.duplicated().sum()
```

```python
#Handling missing values
```

```python
df_olympics.isnull().sum() / len(df_olympics) * 100
```

```python
df_olympics.skew(numeric_only=True)
```

```python
df_olympics = df_olympics.fillna(value={'Age':df_olympics['Age'].median(),
                                       'Height':df_olympics['Height'].mean(),
                                       'Weight':df_olympics['Weight'].median(),
                                       'Medal':'No Medal',
                                       'Region':'Unknown'})
df_olympics.isnull().sum()
```

```python
###Outlier detection and treatment
```

```python
numeric_data = df_olympics.select_dtypes(include=np.number)
```

```python
for col in numeric_data.columns:
    Q1 = df_olympics[col].quantile(0.25)
    Q2 = df_olympics[col].quantile(0.50)
    Q3 = df_olympics[col].quantile(0.75)
    IQR = Q3-Q1
    LW = Q1 - (1.5 * IQR)
    UW = Q3 + (1.5 * IQR)
    outliers = ((df_olympics[col] < LW) | (df_olympics[col] > UW))
    num_outliers = outliers.sum()
    total_rows = len(df_olympics)
    if num_outliers > 0:
        print(f"{col} has {(num_outliers/total_rows)*100:.2f}% outliers")
        df_olympics.loc[(df_olympics[col] < LW) | (df_olympics[col] > UW), col] = Q2
        print(f"Treated outliers for {col} successfully")
    else:
        print(f"{col} does not have any outliers")
    #outlier treatment
    
```

```python
##Find out the medal distribution for top countries
```

```python
subset_df=df_olympics[df_olympics['Medal'] != 'No Medal'][['Region','Event','Year','Medal']].drop_duplicates()
top_regions = subset_df['Region'].value_counts().head(10).index
filtered_df=subset_df[subset_df['Region'].isin(top_regions)]
filtered_df
```

```python
medals_data=filtered_df.groupby(['Region','Medal']).size().unstack()
medals_data.sort_values(by=['Gold','Silver','Bronze'],ascending=False,inplace=True)
medals_data.plot(kind='bar',stacked=True,figsize=(8,5),colormap='viridis')
plt.title('Medal beakdown for top 10 countries')
plt.xlabel('Countries')
plt.ylabel('Medal count')
plt.show()
```
